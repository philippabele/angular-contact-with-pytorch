```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv('../data/iris.csv')
fields = ["sepal.length","sepal.width","petal.length","petal.width"]
```

Get an overview of the data and identify any potential issues or outliers.

```python
# see how the data is distributed and identify any potential skewness or patterns
display(iris[fields].describe())
iris[fields].hist()
plt.show()

# boxplot for identifying outliers and comparing the distribution of the data between different groups or categories
iris[fields].plot(kind='box')
plt.show()
```

Identify relationships between the variables.

```python
# pairplot
sns.pairplot(iris, hue='variety', vars=fields)

# correlation and covariance
df = pd.DataFrame(iris, columns=fields)
cor = df.corr()
cov = df.cov()

fig, ax = plt.subplots(1, 2, figsize=(20, 10))
sns.heatmap(cor, annot=True, ax=ax[0])
ax[0].set_title('Correlation Heatmap');
sns.heatmap(cov, annot=True, ax=ax[1])
ax[1].set_title('Covariance Heatmap');

# cross-correlation
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].xcorr(iris["sepal.length"], iris["sepal.width"], usevlines=True, maxlags=50, normed=True, lw=2)
ax[0].grid(True)
ax[0].set_title('Cross-correlation between sepal length and width');
ax[1].xcorr(iris["petal.length"], iris["petal.width"], usevlines=True, maxlags=50, normed=True, lw=2)
ax[1].grid(True)
ax[1].set_title('Cross-correlation between petal length and width');
```
